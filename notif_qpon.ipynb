```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time
import requests
from datetime import datetime  # Untuk menambahkan waktu

# Token bot Telegram Anda
TELEGRAM_TOKEN = "8165550838:AAFyNjWpsXihvR6Fp6rhOPpuPRpl748nvBg"  # Ganti dengan token bot Anda
CHAT_ID = "1113128625"  # Ganti dengan Chat ID Anda

# URL target
url = "https://qpon-h5-id.qponmobile.com/platform/flashDeal"

# Fungsi untuk mengirim pesan ke Telegram
def send_telegram_message(message):
    try:
        # Endpoint untuk mengirim pesan
        telegram_url = f"https://api.telegram.org/bot{TELEGRAM_TOKEN}/sendMessage"
        payload = {
            "chat_id": CHAT_ID,
            "text": message,
        }
        response = requests.post(telegram_url, data=payload)
        
        # Periksa status pengiriman
        if response.status_code == 200:
            print("Pesan berhasil dikirim ke Telegram!")
        else:
            print(f"Gagal mengirim pesan ke Telegram. Status code: {response.status_code}")
    except Exception as e:
        print(f"Terjadi kesalahan saat mengirim pesan Telegram: {e}")

# Fungsi untuk mengecek warna teks "96"
def check_text_color():
    try:
        # Konfigurasi Chrome untuk berjalan tanpa GUI (headless)
        chrome_options = Options()
        chrome_options.add_argument("--headless")  # Jalankan tanpa jendela browser
        chrome_options.add_argument("--disable-gpu")  # Optimasi untuk headless mode
        chrome_options.add_argument("--no-sandbox")  # Hindari masalah akses root (opsional)

        # Inisialisasi driver Selenium
        service = Service("E:\Project\chromedriver-win64\chromedriver.exe")  # Ganti dengan path ke Chromedriver Anda
        driver = webdriver.Chrome(service=service, options=chrome_options)
        
        # Buka halaman target
        driver.get(url)
        
        # Tunggu beberapa detik agar halaman selesai dimuat
        time.sleep(5)

        # Cari elemen dengan teks "96"
        element = driver.find_element(By.XPATH, "//*[text()='-96']")

        # Ambil warna teks (CSS property 'color')
        color = element.value_of_css_property("color")
        current_time = datetime.now().strftime("%H:%M:%S")  # Format waktu jam:menit:detik


        # Konversi warna ke format RGBA
        if color == "rgba(0, 0, 0, 0.4)":  # Warna untuk habis ()
            print(f"[{current_time}] Stok kosong. Tidak ada perubahan.")
        # elif color == "rgba(221, 8, 102, 1)":  # Warna untuk restok ()
        #     print(f"[{current_time}] Stok tersedia! Mengirim notifikasi...")
        #     send_telegram_message(" QPON ON  QPON ON  : " + url)
        else:
            print(f"[{current_time}]  Stok tersedia! Mengirim notifikasi...")
            send_telegram_message(" QPON ON  QPON ON  : " + url)
        # Tutup driver
        driver.quit()
    except Exception as e:
        print(f"[{datetime.now().strftime('%H:%M:%S')}] Terjadi kesalahan: {e}")

# Loop untuk pengecekan berkala setiap 30 detik
print("Mengecek stok...")
try:
    while True:
        check_text_color()
        time.sleep(15)  # Tunggu 30 detik sebelum pengecekan berikutnya
except KeyboardInterrupt:
    print("Program dihentikan secara manual.")
```
